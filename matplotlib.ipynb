```python
import matplotlib.pyplot as mp
```

```python
x=[1,2,3,4,5,6]
y=[36,95,87,32,65,89]
y1=[33,98,86,100,54,65]
y2=[65,95,87,32,65,48]
y3=[1,2,3,4,5,6]
mp.plot(x,y)
mp.show()
```

```python
mp.subplot(2,2,1)#here 2 is rows 3 is columns 1 is graph number
mp.suptitle('Data')
mp.title('graph-1')
mp.xlabel('x-axis')
mp.ylabel('y-axis')
mp.plot(x,y)

mp.subplot(2,2,2)
mp.title('graph-2')
mp.xlabel('x-axis')
mp.ylabel('y-axis')
mp.plot(x,y1)
mp.show()
```

```python
mp.subplot(2,2,3)
mp.suptitle('Data2')
mp.title('graph-3')
mp.xlabel('x-axis')
mp.ylabel('y-axis')
mp.plot(x,y2)
mp.subplot(2,2,4)
mp.title('graph-4')
mp.xlabel('x-axis')
mp.ylabel('y-axis')
mp.plot(x,y3)
mp.show()
```
